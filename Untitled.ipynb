```python
import pandas as pd

# Load the dataset
df = pd.read_csv('senate_stock_discosures.csv')

# Convert the transaction date to datetime format for proper time-series analysis
df['transaction_date'] = pd.to_datetime(df['transaction_date'])

# Display the top 5 senators by transaction volume
print("--- Top 5 Senators by Transaction Volume ---")
print(df['last_name'].value_counts().head(5))

# Display the top 5 most traded tickers
print("\n--- Top 5 Traded Tickers ---")
print(df['ticker'].value_counts().head(5))

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configure general aesthetics for the plots
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 6))

# Plot 1: Top 10 Senators by transaction count
plt.subplot(1, 2, 1)
top_senators = df['last_name'].value_counts().head(10)
sns.barplot(x=top_senators.values, y=top_senators.index, hue=top_senators.index, legend=False, palette='viridis')
plt.title('Top 10 Senators by Transaction Volume', fontsize=14)
plt.xlabel('Number of Transactions', fontsize=12)
plt.ylabel('Senator Last Name', fontsize=12)

# Plot 2: Top 10 Tickers by transaction count
plt.subplot(1, 2, 2)
top_tickers = df['ticker'].value_counts().head(10)
sns.barplot(x=top_tickers.values, y=top_tickers.index, hue=top_tickers.index, legend=False, palette='magma')
plt.title('Top 10 Most Traded Stocks (Tickers)', fontsize=14)
plt.xlabel('Number of Transactions', fontsize=12)
plt.ylabel('Ticker Symbol', fontsize=12)

# Ensure layout is neat and display the plots
plt.tight_layout()
plt.show()

```

```python
import scipy.stats as stats

# Extract the upper bounds of asset values for Purchase and Sale transactions
purchases = df[df['transaction'].str.contains('Purchase', na=False)]['asset_value_high']
sales = df[df['transaction'].str.contains('Sale', na=False)]['asset_value_high']

# Perform Welch's t-test (assuming unequal variances) to compare investment sizes
t_stat, p_value = stats.ttest_ind(purchases, sales, equal_var=False)

# Print the statistical results
print("--- Hypothesis Test Results ---")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")
print("-" * 30)

# Evaluate the null hypothesis
if p_value < 0.05:
    print("Conclusion: Reject H0")
    print("There is a statistically significant difference between the investment amounts of purchase and sale transactions.")
else:
    print("Conclusion: Fail to reject H0")
    print("There is no statistically significant difference between the investment amounts of purchase and sale transactions.")
    
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# This function checks if a senator is putting all their money in one place.
# 1.0 = All-in on one stock. 0.0 = Fully diversified.
def calculate_hhi(group):
    if group.sum() == 0:
        return 0
    # Weights are the percentage of each stock in the senator's portfolio
    weights = group / group.sum()
    return (weights**2).sum()


# We group by senator and ticker to see the total amount spent per stock
portfolio_totals = df.groupby(['last_name', 'ticker'])['asset_value_high'].sum().reset_index()


# Calculate the focus score (HHI) for every senator.
hhi_analysis = portfolio_totals.groupby('last_name')['asset_value_high'].apply(calculate_hhi).reset_index()
hhi_analysis.columns = ['last_name', 'focus_score']

# Sort to find the top 10 most 'focused' (non-diversified) traders
top_focus_senators = hhi_analysis.sort_values(by='focus_score', ascending=False).head(10)

#PRINT RESULTS
print("--- TOP 10 SENATORS WITH FOCUSED PORTFOLIOS ---")
print(top_focus_senators)

# for VISUALIZATION 
plt.figure(figsize=(12, 6))

# We assign 'last_name' to 'hue' to avoid the library warning
sns.barplot(
    data=top_focus_senators, 
    x='focus_score', 
    y='last_name', 
    hue='last_name', 
    palette='magma', 
    legend=False
)

plt.title('Portfolio Focus Level (HHI Index) by Senator', fontsize=14)
plt.xlabel('Focus Score (1.0 = All eggs in one basket)', fontsize=12)
plt.ylabel('Senator Name', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Selecting the primary numerical features for the anomaly detection task
features = ['asset_value_low', 'asset_value_high'] 

# Dropping missing values to ensure a clean feature matrix for the model
X = df[features].dropna() 

# Initializing the Isolation Forest with a 5% contamination threshold and fitting the data
model = IsolationForest(contamination=0.05, random_state=42)
model.fit(X)

# Generating predictions where -1 represents an outlier and 1 represents an inlier
X['anomaly_status'] = model.predict(X)

# Filtering the dataset to isolate detected anomalies for further analysis
detected_anomalies = X[X['anomaly_status'] == -1]

# Visualizing the outliers using a scatter plot; anomalies are highlighted in red
plt.figure(figsize=(10, 6))
sns.scatterplot(x='asset_value_low', y='asset_value_high', hue='anomaly_status', palette={1: 'blue', -1: 'red'}, data=X)
plt.title('Anomaly Detection Analysis in Financial Transactions')
plt.show()

# Displaying the initial set of suspicious transactions
print("Top 5 Detected Anomalous Transactions:")

# Mapping flagged indices back to the original dataset to identify specific individuals and tickers
flagged_trades = df.loc[detected_anomalies.index, ['first_name', 'last_name', 'ticker', 'transaction_date', 'asset_value_high']]
print("Summary of Identified Anomalous Trades and Corresponding Subjects:")
print(flagged_trades.head(10))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Calculating the estimated transaction volume as the mean of the reported asset value range
df['estimated_volume'] = (df['asset_value_low'] + df['asset_value_high']) / 2

# Quantifying ticker popularity by calculating transaction frequency across the dataset
ticker_frequencies = df['ticker'].value_counts().reset_index()
ticker_frequencies.columns = ['ticker', 'ticker_popularity']

# Integrating popularity metrics into the primary dataframe for feature engineering
if 'ticker_popularity' not in df.columns:
    df = df.merge(ticker_frequencies, on='ticker', how='left')
else:
    df['ticker_popularity'] = df['ticker'].map(ticker_frequencies.set_index('ticker')['ticker_popularity'])

# Defining features for the anomaly detection model: volume vs. popularity
feature_selection = ['estimated_volume', 'ticker_popularity']
X_processed = df[feature_selection].dropna()

# Implementing Isolation Forest with a 2% contamination threshold to isolate extreme outliers
anomaly_detector = IsolationForest(contamination=0.02, random_state=42)
anomaly_detector.fit(X_processed)
X_processed['is_anomaly'] = anomaly_detector.predict(X_processed)

# Filtering the results to isolate identified anomalies
detected_outliers = X_processed[X_processed['is_anomaly'] == -1]

# Extracting relevant metadata for the flagged transactions to analyze suspect behavior
flagged_trades = df.loc[detected_outliers.index, ['first_name', 'last_name', 'ticker', 'transaction_date', 'estimated_volume', 'ticker_popularity']]

# Sorting to prioritize trades with low frequency and high volume (potential information asymmetry)
prioritized_results = flagged_trades.sort_values(by=['ticker_popularity', 'estimated_volume'], ascending=[True, False])

# Visualizing the inverse relationship between ticker popularity and transaction volume
# High-volume trades in obscure tickers are highlighted as anomalies
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='ticker_popularity', 
    y='estimated_volume', 
    hue=X_processed['is_anomaly'], 
    palette={1: 'blue', -1: 'red'}, 
    data=df.loc[X_processed.index]
)
plt.title('Identification of Significant Trades in Low-Popularity Tickers')
plt.xlabel('Ticker Popularity (Transaction Frequency)')
plt.ylabel('Estimated Transaction Volume')
plt.show()

print("Targeted transactions exhibiting high probability of information asymmetry:")
print(prioritized_results.head(15))
```

```python

import yfinance as yf
import pandas as pd
import warnings

# Suppressing warnings to maintain a clean output environment
warnings.filterwarnings('ignore')

def calculate_post_trade_performance(ticker, trade_date_str):
    """
    Retrieves historical price data to calculate the 30-day return following a specific transaction.
    Returns the percentage change or None if data retrieval fails.
    """
    try:
        start_date = pd.to_datetime(trade_date_str)
        # Extending the window to 40 days to ensure we capture enough trading days for a 30-day index
        end_date = start_date + pd.Timedelta(days=40)
        
        # progress=False hides the download bars for a cleaner console log
        stock_data = yf.download(ticker, start=start_date, end=end_date, progress=False)
        
        if stock_data is None or stock_data.empty:
            return None
            
        # Extracting the opening price on the day of (or immediately following) the trade
        price_at_start = float(stock_data['Open'].values[0])
        
        # Locating the price at the 30th trading day or the closest available point
        target_index = 30 if len(stock_data) > 30 else len(stock_data) - 1
        price_at_end = float(stock_data['Close'].values[target_index])
        
        return float(((price_at_end - price_at_start) / price_at_start) * 100)
    except Exception:
        return None

print("Initiating silent data retrieval... Processing identified trades.")

# Iterating through prioritized suspects to quantify post-trade performance
yield_metrics = []
for index, row in prioritized_results.iterrows():
    yield_val = calculate_post_trade_performance(row['ticker'], row['transaction_date'])
    yield_metrics.append(yield_val)

# Integrating performance metrics into the dataset and filtering null entries
prioritized_results['monthly_performance'] = yield_metrics
filtered_outcomes = prioritized_results.dropna(subset=['monthly_performance'])

# Displaying initial performance summary
performance_summary = filtered_outcomes[['first_name', 'last_name', 'ticker', 'transaction_date', 'monthly_performance']]
print("\n--- QUANTIFIED 30-DAY PRICE PERFORMANCE ---")
print(performance_summary.sort_values(by='monthly_performance', ascending=False).to_string(index=False))

# Mapping the original 'transaction' type back to the results for outcome categorization
filtered_outcomes['transaction_type'] = df.loc[filtered_outcomes.index, 'transaction']

def categorize_market_outcome(row):
    """
    Analyzes the correlation between trade direction and subsequent market movement 
    to identify potential information asymmetry or risk avoidance.
    """
    performance = row['monthly_performance']
    trade = str(row['transaction_type']).lower()
    
    if 'purchase' in trade or 'buy' in trade:
        if performance > 0:
            return f"ALPHA_GAIN (+{performance:.1f}%)"
        else:
            return f"UNDERPERFORMANCE ({performance:.1f}%)"
    elif 'sale' in trade or 'sell' in trade:
        if performance > 0:
            return f"OPPORTUNITY_COST (+{performance:.1f}%)"
        else:
            return f"LOSS_AVOIDANCE ({performance:.1f}%)"
    else:
        return "UNDEFINED"

# Applying outcome categorization to evaluate trade significance
filtered_outcomes['outcome_analysis'] = filtered_outcomes.apply(categorize_market_outcome, axis=1)

# Final analytical output: A comprehensive overview of identified information asymmetry
final_analysis_report = filtered_outcomes[['first_name', 'last_name', 'ticker', 'transaction_date', 'transaction_type', 'monthly_performance', 'outcome_analysis']]
print("\n--- FINAL ANALYSIS: INFORMATION ASYMMETRY AND MARKET IMPACT ---")
print(final_analysis_report.sort_values(by='monthly_performance', ascending=False).to_string(index=False))
```

```python
import yfinance as yf
import pandas as pd
import warnings

warnings.filterwarnings('ignore')

# Utilizing the S&P 500 ETF (SPY) as the primary market benchmark for normalization
MARKET_BENCHMARK = "SPY"

def fetch_benchmark_performance(trade_date_str):
    """
    Retrieves the benchmark (SPY) performance over a 30-day horizon post-trade.
    Used to isolate idiosyncratic returns from systematic market movements.
    """
    try:
        start_date = pd.to_datetime(trade_date_str)
        end_date = start_date + pd.Timedelta(days=40)
        
        benchmark_df = yf.download(MARKET_BENCHMARK, start=start_date, end=end_date, progress=False)
        
        if benchmark_df is None or benchmark_df.empty:
            return 0.0
            
        initial_price = float(benchmark_df['Open'].values[0])
        target_idx = 30 if len(benchmark_df) > 30 else len(benchmark_df) - 1
        final_price = float(benchmark_df['Close'].values[target_idx])
        
        return float(((final_price - initial_price) / initial_price) * 100)
    except Exception:
        return 0.0

print("Retrieving benchmark performance metrics for S&P 500...")

# Optimizing runtime by fetching data only for distinct transaction dates
distinct_trade_dates = filtered_outcomes['transaction_date'].unique()
benchmark_yield_map = {date: fetch_benchmark_performance(date) for date in distinct_trade_dates}

# Mapping benchmark returns to the primary dataset
filtered_outcomes['benchmark_30d_yield'] = filtered_outcomes['transaction_date'].map(benchmark_yield_map)

# Quantifying Alpha: The excess return generated over the market benchmark
filtered_outcomes['alpha_generation'] = filtered_outcomes['monthly_performance'] - filtered_outcomes['benchmark_30d_yield']

def analyze_alpha_significance(row):
    """
    Categorizes the trade impact based on the generated Alpha and trade direction.
    Identifies high-conviction signals where the asset decoupled from the broader market.
    """
    alpha = row['alpha_generation']
    trade_dir = str(row['transaction_type']).lower()
    
    if 'purchase' in trade_dir or 'buy' in trade_dir:
        if alpha > 0:
            return f"ALPHA_PURCHASE (Outperformed benchmark by {alpha:.1f}%)"
        else:
            return f"LAGGING_ASSET (Underperformed benchmark by {abs(alpha):.1f}%)"
    elif 'sale' in trade_dir or 'sell' in trade_dir:
        if alpha < 0:
            return f"STRATEGIC_DIVESTMENT (Avoided {abs(alpha):.1f}% decline relative to market)"
        else:
            return f"OPPORTUNITY_LOSS (Asset outperformed benchmark by {alpha:.1f}% post-sale)"
    else:
        return "UNDEFINED"

filtered_outcomes['alpha_impact_assessment'] = filtered_outcomes.apply(analyze_alpha_significance, axis=1)

# Compiling the final performance report with localized alpha metrics
alpha_analysis_report = filtered_outcomes[[
    'first_name', 'last_name', 'ticker', 'transaction_type', 
    'monthly_performance', 'benchmark_30d_yield', 'alpha_generation', 'alpha_impact_assessment'
]]

print("\n--- BENCHMARK-ADJUSTED RELATIVE PERFORMANCE (ALPHA ANALYSIS) ---")
print(alpha_analysis_report.sort_values(by='alpha_generation', ascending=False).to_string(index=False))
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')

# Isolating high-conviction exits characterized by significant negative alpha post-sale
strategic_exits = filtered_outcomes[filtered_outcomes['alpha_impact_assessment'].str.contains("STRATEGIC_DIVESTMENT")].copy()

def identify_crash_event(ticker, exit_date_str):
    """
    Analyzes the 40-day window following an exit to pinpoint the most severe single-day volatility.
    Calculates the proximity of the trade to the market crash and the magnitude of the drawdown.
    """
    try:
        execution_date = pd.to_datetime(exit_date_str)
        # Defining a 40-day observation window to capture delayed market reactions
        window_end = execution_date + pd.Timedelta(days=40)
        
        price_history = yf.download(ticker, start=execution_date, end=window_end, progress=False)
        if price_history is None or price_history.empty or len(price_history) < 5:
            return None, None
            
        # Normalizing column access to handle potential MultiIndex returns from yfinance
        if isinstance(price_history.columns, pd.MultiIndex):
            close_series = price_history['Close'][ticker]
        else:
            close_series = price_history['Close']
            
        # Computing daily percentage returns to identify localized minima
        daily_returns = close_series.pct_change() * 100
        
        # Locating the date and magnitude of the maximum intraday drop within the window
        worst_performing_day = daily_returns.idxmin()
        max_drawdown = float(daily_returns.min())
        
        # Calculating the lead time between the trade execution and the identified crash event
        days_to_volatility = (worst_performing_day - execution_date).days
        
        return days_to_volatility, max_drawdown
    except Exception:
        return None, None

print("Scanning strategic divestments for subsequent crash events and volatility proximity...\n")

event_audit_logs = []
for index, row in strategic_exits.iterrows():
    days_to_event, magnitude = identify_crash_event(row['ticker'], row['transaction_date'])
    
    if days_to_event is not None:
        # Standardizing date presentation for the final report
        if isinstance(row['transaction_date'], pd.Timestamp):
            date_label = row['transaction_date'].strftime('%Y-%m-%d')
        else:
            date_label = str(row['transaction_date']).split(' ')[0]
            
        event_audit_logs.append({
            'Subject': f"{row['first_name']} {row['last_name']}",
            'Ticker': row['ticker'],
            'Exit_Date': date_label,
            'Days_to_Crash': int(days_to_event),
            'Event_Magnitude': f"{magnitude:.1f}%",
            'Total_Alpha_Avoided': f"{row['alpha_generation']:.1f}%"
        })

# Generating the structured audit report for anomalous exit timing
audit_report = pd.DataFrame(event_audit_logs)

if not audit_report.empty:
    # Prioritizing results by proximity to the event to highlight potential information asymmetry
    print(audit_report.sort_values(by=['Days_to_Crash', 'Total_Alpha_Avoided'], ascending=[True, True]).to_string(index=False))
else:
    print("No statistically significant post-trade volatility events detected within the sample.")
    
```

```python
import pandas as pd
from itertools import combinations
import warnings

warnings.filterwarnings('ignore')

print("Analyzing 'Herd Behavior' (Co-trading Network) among subjects...\n")

# Sorting the processed dataset chronologically to establish an execution sequence
sorted_outcomes = filtered_outcomes.sort_values('transaction_date')

co_trade_links = []

# Grouping activity by ticker to analyze clustering within specific assets
for ticker, trade_cluster in sorted_outcomes.groupby('ticker'):
    # Extracting trading records for all subjects involved with the current ticker
    trading_subjects = trade_cluster[['first_name', 'last_name', 'transaction_date', 'transaction_type', 'alpha_generation']].to_dict('records')
    
    # Generating pairwise combinations to identify subjects trading the same asset
    for subject_a, subject_b in combinations(trading_subjects, 2):
        name_a = f"{subject_a['first_name']} {subject_a['last_name']}"
        name_b = f"{subject_b['first_name']} {subject_b['last_name']}"
        
        if name_a != name_b:
            # Quantifying the temporal proximity between the two execution dates
            execution_gap = abs((pd.to_datetime(subject_a['transaction_date']) - pd.to_datetime(subject_b['transaction_date'])).days)
            
            # Filtering for trades occurring within a 90-day window (one fiscal quarter) 
            # to establish a statistically significant co-trading relationship
            if execution_gap <= 90:
                co_trade_links.append({
                    'Subject_A': name_a,
                    'Subject_B': name_b,
                    'Shared_Ticker': ticker,
                    'Execution_Gap': execution_gap,
                    'Trade_A': str(subject_a['transaction_type']).split(' ')[0],
                    'Trade_B': str(subject_b['transaction_type']).split(' ')[0]
                })

network_data = pd.DataFrame(co_trade_links)

if not network_data.empty:
    # Aggregating the frequency of shared trades to identify high-conviction clusters
    frequency_summary = network_data.groupby(['Subject_A', 'Subject_B']).size().reset_index(name='Shared_Trade_Count')
    frequency_summary = frequency_summary.sort_values(by='Shared_Trade_Count', ascending=False)
    
    print("--- HIGHEST CO-TRADING FREQUENCY BY PAIR ---")
    print(frequency_summary.to_string(index=False))
    
    print("\n--- DETAILED CO-TRADING LOG (SORTED BY TEMPORAL PROXIMITY) ---")
    print(network_data.sort_values(by='Execution_Gap').head(15).to_string(index=False))
else:
    print("No suspicious network activity or herd behavior detected within the 90-day threshold.")




    
    
```
